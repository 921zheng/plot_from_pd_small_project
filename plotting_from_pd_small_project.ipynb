```python
df["Price"]=df["Price"].str.replace('[\$\,\.]','').astype(int)/100
df
```

```python
#add sale_date
df["Sale Date"]=pd.date_range("1/1/2023", periods=len(df["Make"]))
df
```

```python
#add total_sale column by cumsum()
df["Total Sale"]=df["Price"].cumsum()
df
```

```python
#plot it
df.plot(x="Sale Date", y="Total Sale")
```
